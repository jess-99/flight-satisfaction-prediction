# Normalization
## using minimax algorithm

```python
data1=data
```

```python
data1.head()
```

```python
data1["Age"]=((data1["Age"]-data1["Age"].min())/(data1["Age"].max()-data1["Age"].min()))
```

```python
data1["Age"].head()
```

```python
data1["Flight Distance"]=((data1["Flight Distance"]-data1["Flight Distance"].min())/(data1["Flight Distance"].max()-data1["Flight Distance"].min()))
```

```python
data1["Flight Distance"].head()
```

```python
data1["Seat comfort"]=((data1["Seat comfort"]-data1["Seat comfort"].min())/(data1["Seat comfort"].max()-data1["Seat comfort"].min()))
```

```python
data1["Seat comfort"].head()
```

```python
data1["Departure/Arrival time convenient"]=((data1["Departure/Arrival time convenient"]-data1["Departure/Arrival time convenient"].min())/(data1["Departure/Arrival time convenient"].max()-data1["Departure/Arrival time convenient"].min()))
```

```python
data1["Departure/Arrival time convenient"].head()
```

```python
data1.head()
```

```python
data1.describe()
```

```python
le = preprocessing.LabelEncoder()
```

```python
data1['satisfaction_v2'] = le.fit_transform(data1['satisfaction_v2'])
data1['Gender'] = le.fit_transform(data1['Gender'])
data1['Customer Type'] = le.fit_transform(data1['Customer Type'])
data1['Type of Travel'] = le.fit_transform(data1['Type of Travel'])
data1['Class'] = le.fit_transform(data1['Class'])
```

```python
data1.head()
```

## naive bayes algorithm

```python
features = data1.loc[:, 'Gender':'Departure/Arrival time convenient']
label = data1['satisfaction_v2']
```

```python
features
```

```python
label
```

```python
from sklearn.naive_bayes import CategoricalNB
from sklearn import metrics
import numpy as np
```

```python
model = CategoricalNB()
```

```python
res = model.fit(features, label)
```

```python
res
```

```python
predict=model.predict(features)
predict
```

```python
actual=np.asarray(label)
actual
```

```python
data1.head()
```

### accuracy

```python
print('accuracy = ', metrics.accuracy_score(label,predict ))
```

### confusion matrix

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
matrix = confusion_matrix(actual,predict, labels=[1,0])
print('Confusion matrix : \n',matrix)
```

```python
tp, fn, fp, tn = confusion_matrix(actual,predict,labels=[1,0]).reshape(-1)
print('Outcome values : \n','TP: FN: FP: TN: \n', tp, fn, fp, tn)
```

```python
matrix = classification_report(actual,predict,labels=[1,0])
print('Classification report : \n',matrix)
```

```python
d_predict=predict
d_actual=actual
data2 = {'actual':    d_actual,
        'Predict': d_predict
        }
df = pd.DataFrame(data2, columns=['actual','Predict'])
confusion_matrix = pd.crosstab(df['actual'], df['Predict'], rownames=['Actual'], colnames=['Predict'], margins = True)

sn.heatmap(confusion_matrix, annot=True)
plt.show()
```
